```python
for img in get_image_files(foodPath):
    if labelA in str(img):
        img.rename(f"{img.parent}/{labelA}-{img.name}")
    elif labelB in str(img):
        img.rename(f"{img.parent}/{labelB}-{img.name}")
    elif labelC in str(img):
        img.rename(f"{img.parent}/{labelC}-{img.name}")
    else:
        os.remove(img)

```

```python
dls = ImageDataLoaders.from_name_func(foodPath, get_image_files(foodPath), 
                                        valid_pct=0.2, seed=42, label_func=GetLabel, item_tfms=Resize(224))

learn = vision.learner(dls, resnet34, metrics = accuracy, pretrained= True)
```

```python
learn.fine_tune(epochs=10)
learn.export('/kaggle/working/export1.pkl')
#dls.train.show_batch()
```
